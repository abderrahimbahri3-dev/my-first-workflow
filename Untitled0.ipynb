```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models

# 1. تحميل شبكة ResNet50 جاهزة (بدون الطبقة الأخيرة)
# include_top=False يعني أننا سنحذف "الرأس" القديم للشبكة لنضع رأسنا الخاص
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 2. تجميد الطبقات (Freezing)
# نخبر الموديل ألا يغير الأوزان التي تعلمها من قبل في الطبقات الأساسية
base_model.trainable = False

# 3. بناء الهيكل الجديد (Our Custom Head)
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(), # تحويل البيانات المعقدة إلى ناقل بسيط
    layers.Flatten(),                # مسح البيانات لجعلها خطية
    layers.Dense(128, activation='relu'), # طبقة مخفية للتعلم العميق
    layers.Dropout(0.2),             # لمنع الموديل من "الحفظ" الزائد (Overfitting)
    layers.Dense(1, activation='sigmoid') # الطبقة الأخيرة: 1 لأننا نملك صنفين فقط (0 أو 1)
])

# 4. تجميع الموديل (Compile)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary() # لعرض ملخص لكل طبقات الموديل
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. تحديد حجم الصورة المطلوب لـ ResNet50
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# 2. إعداد مولد بيانات التدريب (مع عمل Augmentation)
train_datagen = ImageDataGenerator(
    rescale=1./255,          # تحويل القيم لـ [0,1]
    shear_range=0.2,         # قص خفيف
    zoom_range=0.2,          # زوم خفيف
    horizontal_flip=True     # قلب الصورة أفقياً
)

# 3. إعداد مولد بيانات الاختبار (فقط Rescale بدون تلاعب)
test_datagen = ImageDataGenerator(rescale=1./255)

# 4. تحميل الصور فعلياً من المجلدات
train_generator = train_datagen.flow_from_directory(
    '/content/dataset/chest_xray/train',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'      # لأننا نصنف نوعين فقط
)

test_generator = test_datagen.flow_from_directory(
    '/content/dataset/chest_xray/test',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

print("تم تجهيز المولدات بنجاح!")
```
